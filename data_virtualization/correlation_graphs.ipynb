# Correlation Graphs:

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips") #loading data from seaborn library
df = tips.copy() #Creating a new variable to work on dataset
df.head() #To see first 5 elements in dataset
```

**Scatterplot:**

```python
sns.scatterplot(x = "total_bill", 
                y = "tip", 
                data = df);
```

**Cross Process in Correlation Plot:**

```python
#We're used hue settings for adding categorical variable.
sns.scatterplot(x = "total_bill", 
                y = "tip", 
                hue = "time", 
                data = df);
```

**Adding style setting for seeing graph more easier, and understandable:**

```python
sns.scatterplot(x = "total_bill", 
                y = "tip", 
                hue = "time", 
                style = "time",
                data = df);
```

```python
sns.scatterplot(x = "total_bill", 
                y = "tip", 
                hue = "day", 
                style = "time",
                data = df);
```

**Viewing correlation graph with points of variable**

```python
sns.scatterplot(x="total_bill", 
                y = "tip", 
                size = "size", 
                data = df);
```

**Showing the linear relationships:**

```python
#To draw selected variable
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x = "total_bill",
           y = "tip",
           data = df);
```

```python
sns.lmplot(x = "total_bill",
           y = "tip",
           hue = "smoker",
           data = df);
```

**Dividing the correlation graph:**

```python
#Quote: The common variable is `time` in this graph:
sns.lmplot(x = "total_bill", 
           y = "tip", 
           hue = "smoker", 
           col = "time", 
           data = df);
```

```python
#Quote: The common variable is `time`  & `sex `in this graph:
sns.lmplot(x = "total_bill", 
           y = "tip", 
           hue = "smoker", 
           col = "time", 
           row = "sex",
           data = df);
```
