```python
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
df = diamonds.copy()
```

```python
df.head()
```

```python
sns.distplot(df.price, kde= False);
```

```python
sns.distplot(df.price, bins = 1000, kde= False);
```

```python
sns.distplot(df.price);
```

```python
sns.distplot(df.price, hist= False);
```

```python
sns.distplot(df.price, shade = True);
```

# Cross Process in Histogram with Many Variables:

```python
#If we're finding the answer of "how did?" question,
#Genereally we're using this histogram
#With this options:
sns.kdeplot(df.price, shade = True);
```

```python
#Facetgrid is using for graphic to show graphic by divide variables
#Argument of Hue is using for adding dimension
(sns.FacetGrid(df, 
              hue = "cut", 
              height = 5,
              xlim = (0,10000) #To setting X Axis
             )
 .map(sns.kdeplot, "price", shade = True)
 .add_legend() #To add cut categories' info
);
```

```python
sns.catplot(x = "cut", #To give name & values X Axis
            y = "price", #To give name & values Y axis
            hue = "color", #To add colors
            kind = "point", #To see graph with points
            data = df); 
```
