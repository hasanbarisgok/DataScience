```python
import pandas as pd 
```

```python
amount_of_wastes = pd.read_csv("https://raw.githubusercontent.com/hasanbarisgok/DataScience/main/opensource-datas/ibb_ilcelere_gore_atik_miktari.csv", header=0)
df = amount_of_wastes
```

```python
df.head()
```

```python
df.describe().T
```

```python
#Is the dataset has any NaN value?
df.isnull().values.any()
```

```python
#NaN values count in the dataset 
df.isna().sum().sum()
```

```python
#Check NaN values with isnull() function. True = NaN Value, False = Is Not Nan Value
df.isnull()
```

```python
#Giving value for NaNs. All NaN values will be `0` (zero).
df = df.fillna(0)
```

```python
#Is the dataset has any NaN value?
df.isnull().values.any()
```

```python
#The first five elements from the dataset.
df.head()
```

```python
#Removing `Tibbi Atik Miktari` column. We have to give value 1 to axis. `Axis = 1 for Columns, Axis = 0 for Rows`
df = df.drop("Tıbbi Atık Miktarı", axis = 1 )
df.head()
```

```python
#In this code, we're reading another csv. This csv includes all Turkey's Districts. 
df_enlem_boylam = pd.read_csv("https://gist.githubusercontent.com/abdullahoguk/ee03c26a23dca6eda9c480b4967e77b6/raw/cd6774a079b03ad9613de0e78dc9b9cbda2244de/ilce.csv")
#The first element from new dataset.
df_enlem_boylam.head(1)
```

```python
#We're grouped the new dataset, which plate_no (il_plaka) 34. Because Istanbul's plate no is `34`. So, after that, we're reach only Istanbul's plates no.
df_enlem_boylam = df_enlem_boylam[df_enlem_boylam.il_plaka == 34]
```

```python
#We're grouping the data set with only `district_name, it's latitude_value, it's longtidude_value`
df_enlem_boylam = df_enlem_boylam[["ilce_adi","lat","lon"]].reset_index(drop=True)
```

```python
#The changed new dataset is there:
df_enlem_boylam.head()
```

```python
#We're adding latitude values and longtidude values to `df` dataset. This likes concat function. But we're not using concat, because its not 
#necessary for this section for me. The datasets indexes values are equal.(both name types and sorting). So there is no problem. This is
#really easy for us.

df[["lat","lon"]] = df_enlem_boylam[["lat","lon"]]
```

```python
#The first five elements new dataset. (after adding lat & lot elements)
df.head()
```
