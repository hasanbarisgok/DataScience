```python
import numpy as np
import pandas as pd
```

```python
matris = np.random.randint(1,30, size = (5,3))
df = pd.DataFrame(matris, columns = ["variable_1","variable_2","variable_3"])
df
```

```python
df_2 = df + 50
df
```

```python
pd.concat([df,df_2])
```

```python
#If we use the `ignore_index = true` in this
#technique, we can solve the index problem.
#You can see the index problem above this code.
pd.concat([df,df_2], ignore_index=True)
```

**Changing Column Names:**

```python
#If we use the variable.columns technique,
#We can change the column/variable names.
#Here is the example:
df_2.columns = ["variable_1","var_2","new_variable_3"]
```

```python
#Here is the df_2 with new column names.
df_2
```

**Quote:** If you want to join DataFrame, which are columns
name's are different, you'll have a problem. The problem name will be `future_warning`. If the program doesn't give any error, you have many unnecessary columns with Nan Values.


```python
pd.concat([df,df_2])
```

You can solve all problems like above as using that technique:
`join = inner`

Here is the example:

```python
pd.concat([df,df_2], join = "inner")
```
