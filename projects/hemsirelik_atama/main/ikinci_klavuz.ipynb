```python
import pandas as pd
```

```python
ikinci_klavuz = pd.read_csv("https://raw.githubusercontent.com/hasanbarisgok/DataScience/main/projects/hemsirelik_atama/16_11_2021.csv")
```

```python
ikinci_klavuz.describe().T
```

```python
ikinci_klavuz.head()
```

```python
ikinci_klavuz = ikinci_klavuz.drop(columns = ["Unnamed: 5", "1546", "4605"])
```

```python
ikinci_klavuz.head()
```

```python
ikinci_klavuz = ikinci_klavuz.rename(columns = {"16.12.2021" : "HISTORY", "Başakşehir Çam ve Sakura Şehir Hast." : "HOSPITALS", "Hemşire (İstanbul)":"CITIES","669":"QUOATA", "71,07":"MIN",
                                "79,7":"MAX"})
```

```python
ikinci_klavuz.head()
```

```python
regex = r'\((.*?)\)'
```

```python
ikinci_klavuz["CITIES"] = ikinci_klavuz["CITIES"].str.extract(regex)
```

```python
ikinci_klavuz.head()
```

```python
regions = pd.read_csv("https://raw.githubusercontent.com/hasanbarisgok/TurkiyeSehirlerBolgeler/master/csv/sehirler_ve_bolgeler.csv")
regions = regions[["SehirAd","BolgeAd"]]
regions = regions.rename(columns = {"SehirAd":"CITIES","BolgeAd":"REGIONS"})
regions.head()
```

```python
ikinci_klavuz = ikinci_klavuz.merge(regions, how = "inner", on = "CITIES")
ikinci_klavuz.head()
```

```python
ikinci_klavuz.shape
```

```python
ikinci_klavuz["QUOATA"].max()
```

```python
ikinci_klavuz["QUOATA"].min()
```

```python
dogu_anadolu_ikinci_klavuz = ikinci_klavuz.loc[ikinci_klavuz["REGIONS"] == "Doğu Anadolu Bölgesi"]
karadeniz_ikinci_klavuz = ikinci_klavuz.loc[ikinci_klavuz["REGIONS"] == "Karadeniz"]
ege_ikinci_klavuz = ikinci_klavuz.loc[ikinci_klavuz["REGIONS"] == "Ege Bölgesi"]
akdeniz_ikinci_klavuz = ikinci_klavuz.loc[ikinci_klavuz["REGIONS"] == "Akdeniz Bölgesi"]
marmara_ikinci_klavuz = ikinci_klavuz.loc[ikinci_klavuz["REGIONS"] == "Marmara Bölgesi"]
```

```python
dogu_anadolu_ikinci_klavuz.to_csv("dogu_anadolu_ikinci_klavuz.csv",encoding='utf-8-sig')
karadeniz_ikinci_klavuz.to_csv("karadeniz_ikinci_klavuz.csv",encoding='utf-8-sig')
ege_ikinci_klavuz.to_csv("ege_ikinci_klavuz.csv",encoding='utf-8-sig')
akdeniz_ikinci_klavuz.to_csv("akdeniz_ikinci_klavuz.csv",encoding='utf-8-sig')
marmara_ikinci_klavuz.to_csv("marmara_ikinci_klavuz.csv",encoding='utf-8-sig')
```
