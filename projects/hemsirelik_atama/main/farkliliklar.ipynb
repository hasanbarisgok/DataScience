# 1.Veride olup, 2.Veride olmayan hastaneler ve atamalar:

```python
import pandas as pd 
```

```python
df1 = pd.read_csv("https://raw.githubusercontent.com/hasanbarisgok/DataScience/main/projects/hemsirelik_atama/main/result_csvs/birinci_klavuz_tum.csv")
df2 = pd.read_csv("https://raw.githubusercontent.com/hasanbarisgok/DataScience/main/projects/hemsirelik_atama/main/result_csvs/ikinci_klavuz_tum.csv")
```

```python
mask = ~df1['HOSPITALS'].isin(df2['HOSPITALS'])
result = df1[mask]
```

```python
result.head()
```

```python
result = result.drop(columns = ["Unnamed: 0", "VACANT"], axis = 1)
result = result.rename(columns = {"TOTAL QUOATA":"QUOATA","LOWEST POINT":"MIN","HIGHEST POINT":"MAX"})
```

```python
result = result.sort_values(by = ["CITIES", "HOSPITALS"])
```

```python
result = result.reset_index(drop=True)
```

```python
result.head()
```

```python
result.describe().T
```

```python
result.to_csv("sadece_birinci_klavuzdaki_hastaneler.csv", index = False)
```
