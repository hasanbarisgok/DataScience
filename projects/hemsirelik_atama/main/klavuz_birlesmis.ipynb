```python
import pandas as pd
```

```python
df_1 = pd.read_csv("https://raw.githubusercontent.com/hasanbarisgok/DataScience/main/projects/hemsirelik_atama/main/result_csvs/birinci_klavuz_tum.csv")
df_2 = pd.read_csv("https://raw.githubusercontent.com/hasanbarisgok/DataScience/main/projects/hemsirelik_atama/main/result_csvs/ikinci_klavuz_tum.csv")
```

```python
df_1.drop(df_1.filter(regex="Unname"), axis=1, inplace=True)
df_2.drop(df_2.filter(regex="Unname"), axis=1, inplace=True)
```

```python
df_1 = df_1.drop("VACANT",axis=1)
df_1 = df_1.rename(columns={"TOTAL QUOATA":"QUOATA","LOWEST POINT":"MIN","HIGHEST POINT":"MAX"})
```

```python
df_1.head()
```

```python
df_2.head()
```

```python
birlesik_veri = pd.merge(df_1, df_2, left_on=['HOSPITALS','CITIES','REGIONS'], right_on = ['HOSPITALS','CITIES','REGIONS'],suffixes=(' 1.KLAVUZ', ' 2.KLAVUZ'))
```

```python
birlesik_veri = birlesik_veri.sort_values(by="CITIES")
birlesik_veri.head()
```

```python
birlesik_veri.describe().T
```

```python
birlesik_yeni_veri = birlesik_veri[["HOSPITALS","CITIES","QUOATA 1.KLAVUZ","MIN 2.KLAVUZ","MAX 2.KLAVUZ","REGIONS"]]
birlesik_yeni_veri = birlesik_yeni_veri.sort_values(by="CITIES")
birlesik_yeni_veri.head()
```

```python
birlesik_yeni_veri["QUOATA 1.KLAVUZ"] = birlesik_yeni_veri["QUOATA 1.KLAVUZ"] + birlesik_veri["QUOATA 2.KLAVUZ"]
```

```python
birlesik_yeni_veri.head()
```

```python
birlesik_yeni_veri = birlesik_yeni_veri.rename(columns = {"QUOATA 1.KLAVUZ":"QUOATA", "MIN 2.KLAVUZ" : "MIN", "MAX 2.KLAVUZ" : "MAX"})
```

```python
birlesik_yeni_veri = birlesik_yeni_veri.reset_index(drop=True)
```

```python
birlesik_yeni_veri.head()
```

```python
birlesik_yeni_veri.to_csv("birlesik_veri_kesisim_yeni.csv", index = False)
```
